```python
#Splitting the data into x,y variable
x = new_dataset.drop(columns = 'Class',axis = 1)
y = new_dataset["Class"]
```

```python
x
```

```python
y
```

```python
#Importing the train_test_split and LogisticRegression Algorithm
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size =0.8,random_state =2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Creting the logisticRegression  model
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#Accuracy on train data
predict = model.predict(x_train)
```

```python
accuracy = accuracy_score(predict,y_train)
```

```python
accuracy
```

```python
#Accuracy on test data 
x_test_predict = model.predict(x_test)
```

```python
Accuracy = accuracy_score(x_test_predict,y_test)
```
